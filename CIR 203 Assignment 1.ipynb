Exercise 1: Arrays in Financial Analytics (Banking Sector)

```python
import numpy as np
```

```python
# 1. Use NumPy to create a 2D array representing transaction volumes for 4 branches over 6 months.
# Each row represents a branch, and each column a month. (4 marks)

transVolumes = np.array([
                         [100, 250, 320, 400, 570, 620],
                         [706, 870, 909, 1050, 1500, 2000],
                         [650, 570, 460, 350, 290, 1000],
                         [2000, 1050, 1000, 950, 810, 709]
                        ])
```

```python
# 2. Calculate the total transactions per branch. (4 marks)

branches = ['Branch 1', 'Branch 2', 'Branch 3', 'Branch 4']
totalTrans = transVolumes.sum(axis=1)
for add, branch in zip(totalTrans, branches):
    print(f"The total transactions of {branch} is: {add}")
```

```python
# 3. Identify the branch with the highest total transactions. (4 marks)

highTotalTrans = np.argmax(totalTrans)
print(f"{branches[1]} has the highest total transaction of {totalTrans[highTotalTrans]}")
```

```python
# 4. Compute the average monthly transaction volume across all branches. (4 marks)

averageTransac = transVolumes.mean(axis=1)
for average, branch in zip(averageTransac, branches):
    print(f"The average monthly transaction volume of {branch} is: {average}")
```

```python
# 5. Reshape the array to a 3x8 format and explain the implications of reshaping in this context. (4marks)

reshapedArray = transVolumes.reshape(3, 8)
```

```python
print(reshapedArray)
```

Exercise 2: Lists in Logistics Management (Supply Chain Sector)

```python
# 1. Create a list of 10 delivery routes, each represented as a string. (4 marks)

deliveRoutes = list((
                    'Maseno Unversity', 'Kisumu', 'Luanda', 'Nyalenda area', 'Nairobi',
                    'Mombasa', 'Nakuru', 'Kariobangi', 'Kiambu', 'Migori'
                    ))
```

```python
# 2. Append a new route and remove a discontinued one. (4 marks)

deliveRoutes.append('Kilifi')

deliveRoutes.remove(deliveRoutes[7])
```

```python
# 3. Sort the list alphabetically and reverse it. (4 marks)

deliveRoutes.sort(reverse=True)
```

```python
# 4. Count how many routes start with the letter “N”. (4 marks)

count_N = sum(route.startswith('N') for route in deliveRoutes)

print(f"Number of routes satrting with letter 'N' is: {count_N}")
```

```python
# 5. Use list comprehension to create a new list of routes longer than 10 characters. (4 marks)

long_routes = list((route for route in deliveRoutes if len(route) > 10))
print(long_routes)
```

Exercise 3: Tuples in Medical Records (Healthcare Sector)

```python
# 1. Create a tuple for a patient containing: (Name, Age, Blood Pressure, Heart Rate). (4 marks)

patient = tuple(('Steve', 20, '120/80', 72))
```

```python
# 2. Access and print the patient’s age and heart rate. (4 marks)

print(f"{patient[0]} is {patient[1]} years old with {patient[3]} heart rate")
```

```python
# 4. Convert the tuple to a list, update the heart rate, and convert it back to a tuple. (4 marks)

convertedToList = list(patient)
convertedToList[3] = 75
convertedToTuple = tuple(convertedToList)
print(convertedToTuple)
```

```python
# 5. Create a tuple of 5 patients and write code to extract all names. (4 marks)

patients = (
            ('Steve', 20, '120/80', 75),
            ('Elsa', 20, '121/80', 72),
            ('Charles', 21, '122/80', 70),
            ('Sam', 23, '123/80', 77),
            ('Rael', 19, '124/80', 78)
)

for patient in patients:
    name, age, bp, hr = patient
    print(f"{name} is {age} years old having {bp} blood pressure with {hr} heart rate")
```

Exercise 4: Dictionaries in E-Commerce (Retail Sector)

```python
# 1. Create a dictionary with 5 products as keys and their stock quantities as values. (4 marks)

products = dict([
                ('Cat6 cable', 9),
                ('Redmi 13C', 10),
                ('RJ 45 Connectors', 5),
                ('Mouse', 9),
                ('Monitors', 20),
])
```

```python
# 2. Add a new product and update the quantity of an existing one. (4 marks)

products['Keyboard'] = 20
products.update({'Redmi 13C': 20})
```

```python
# 3. Write a function to return products with stock less than 10. (4 marks)

def LessStock():
    return {product: qty for product, qty in products.items() if qty < 10}

LessStock()
```

```python
# 4. Delete a discontinued product and display the updated dictionary. (4 marks)

products.pop('Monitors')
print(products)
```

```python
# 5. Use .items() to loop through and print each product with its quantity. (4 marks)

for product, qty in products.items():
    print(f"{product} has {qty} left")
```
